Este notebook contém a resolução completa do projeto de análise de dados do Banco Fácil, seguindo todas as tarefas especificadas.

```python
# Tarefa 0: Importar bibliotecas e criar as tabelas
# Importando pandas e numpy para manipulação de dados
import pandas as pd
import numpy as np

# Base de transações
transacoes = pd.DataFrame({
    'id_cliente': [1, 2, 1, 3, 2, 1, 3, 2],
    'tipo':       ['Crédito', 'Débito', 'Débito', 'Crédito', 'Crédito', 'Débito', 'Débito', 'Crédito'],
    'categoria':  ['Salário', 'Alimentação', 'Transporte', 'Salário', 'Lazer', 'Alimentação', 'Transporte', 'Salário'],
    'valor':      [3200.0, 150.0, None, 4500.0, 200.0, 80.0, None, 2800.0],
    'canal':      ['App', 'Agência', 'App', 'Internet Banking', 'App', 'Agência', 'App', 'Internet Banking']
})

# Base de cadastro de clientes
clientes = pd.DataFrame({
    'id_cliente': [1, 2, 3, 4],
    'nome':       ['Ana', 'Bruno', 'Carlos', 'Diana'],
    'estado':     ['SP', 'RJ', 'MG', 'SP'],
    'perfil':     ['Premium', 'Básico', 'Intermediário', 'Básico']
})

print("Tabelas criadas com sucesso!")
```

## Tarefa 1 — Conheça os dados

Vamos explorar as duas tabelas usando `shape`, `dtypes`, `head()` e `isna().sum()`.

```python
# Explorando a tabela de transações
print("Tabela de Transações:")
print(f"Shape: {transacoes.shape}")
print(f"Dtypes:\n{transacoes.dtypes}")
print(f"Head:\n{transacoes.head()}")
print(f"Valores nulos:\n{transacoes.isna().sum()}")
print("\n" + "="*50 + "\n")

# Explorando a tabela de clientes
print("Tabela de Clientes:")
print(f"Shape: {clientes.shape}")
print(f"Dtypes:\n{clientes.dtypes}")
print(f"Head:\n{clientes.head()}")
print(f"Valores nulos:\n{clientes.isna().sum()}")

# Respostas:
# - Quantos clientes existem no cadastro? 4
# - Quantas transações existem? 8
# - Alguma coluna possui valores nulos? Sim, a coluna 'valor' na tabela transacoes tem 2 valores nulos.
```

## Tarefa 2 — Limpeza

Preencher os valores nulos da coluna `valor` com a mediana geral da coluna.

```python
# Calculando a mediana da coluna valor (ignorando nulos)
mediana_valor = transacoes['valor'].median()
print(f"Mediana da coluna valor: {mediana_valor}")

# Preenchendo os nulos com a mediana
transacoes['valor'] = transacoes['valor'].fillna(mediana_valor)

# Confirmando que não há mais nulos
print(f"Valores nulos após limpeza:\n{transacoes.isna().sum()}")
print(f"Tabela após limpeza:\n{transacoes}")
```

## Tarefa 3 — Cálculos com NumPy

Criar um array NumPy com os valores da coluna `valor` e calcular soma, média, maior e menor valor.

```python
# Criando array NumPy com os valores
valores_array = np.array(transacoes['valor'])
print(f"Array NumPy: {valores_array}")

# Cálculos
soma_total = np.sum(valores_array)
media = np.mean(valores_array)
maior = np.max(valores_array)
menor = np.min(valores_array)

print(f"Soma total das transações: {soma_total}")
print(f"Valor médio: {media}")
print(f"Maior valor: {maior}")
print(f"Menor valor: {menor}")
```

## Tarefa 4 — Join entre as bases

Fazer inner join e left join, comparar shapes.

```python
# Inner join
df_inner = pd.merge(transacoes, clientes, on='id_cliente', how='inner')
print(f"Shape do inner join: {df_inner.shape}")
print(f"Inner join:\n{df_inner}")

# Left join partindo de clientes
df_left = pd.merge(clientes, transacoes, on='id_cliente', how='left')
print(f"\nShape do left join: {df_left.shape}")
print(f"Left join:\n{df_left}")
```

**Explicação:** Os shapes são diferentes porque o inner join inclui apenas os clientes que têm transações (id_cliente 1,2,3), resultando em 8 linhas (mesmo número de transações). O left join inclui todos os clientes (4), mas adiciona as transações correspondentes, resultando em mais linhas se um cliente tiver múltiplas transações, ou linhas com NaN se não tiver. Aqui, Diana (id 4) não tem transações, então df_left tem 9 linhas (8 transações + 1 cliente sem transação).

## Tarefa 5 — Investigando o left join

Filtrar linhas onde `valor` está nulo no `df_left`.

```python
# Filtrando linhas com valor nulo
clientes_sem_transacao = df_left.loc[df_left['valor'].isna()]
print(f"Clientes sem transação:\n{clientes_sem_transacao}")

# Resposta: A cliente Diana (id 4) consta no cadastro mas nunca realizou uma transação.
```

## Tarefa 6 — Análise por perfil

Usar groupby na coluna `perfil` para calcular soma e média dos valores.

```python
# Groupby por perfil
analise_perfil = df_inner.groupby('perfil')['valor'].agg(['sum', 'mean'])
print(f"Análise por perfil:\n{analise_perfil}")

# Qual perfil teve o maior valor médio?
maior_media = analise_perfil['mean'].idxmax()
print(f"\nPerfil com maior valor médio: {maior_media}")
```

## Tarefa 7 — Salvar o resultado

Salvar o `df_inner` em `base_combinada.csv` e ler de volta.

```python
# Salvando o df_inner
df_inner.to_csv('base_combinada.csv', index=False)
print("Arquivo 'base_combinada.csv' salvo com sucesso!")

# Lendo de volta para confirmar
df_lido = pd.read_csv('base_combinada.csv')
print(f"\nArquivo lido:\n{df_lido}")
print(f"Shape do arquivo lido: {df_lido.shape}")
```
